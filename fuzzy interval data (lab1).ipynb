```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz

data = np.array([[2, 5, 8], [4, 7, 11], [6, 9, 11], [8, 11, 11]])
x = np.linspace(0, 15, 1000)
membership_funcs = [fuzz.trimf(x, interval) for interval in data]
```

```python
for i, mf in enumerate(membership_funcs):
    plt.plot(x, mf, label=f'Interval {i+1}')
plt.title(" Interval ma'lumotlarini fazzyfikatsiyalash")
plt.xlabel('Qiymati')
plt.ylabel('Mansubligi')
plt.legend()
plt.show()
```

```python
input_data = np.array([[0, 5,10], [5, 10, 15], [10, 16, 20]])
output_data = np.array([[0, 0.5,1], [5, 7,10], [10, 12, 15]])

# Kirish va chiqish uchun noravshan mansublik funksiyalarini yarating

input_mfs = [fuzz.trimf(x, interval) for interval in input_data]
output_mfs = [fuzz.trimf(x, interval) for interval in output_data]

# Kirish va chiqish mansublik funksiyalarini taqdim qiling
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
for i, mf in enumerate(input_mfs):
    plt.plot(x, mf, label=f'Input Set {i+1}')

plt.title("Kirish mansublik funktsiyalari")
plt.xlabel('Qiymat')
plt.ylabel('Mansublik')
plt.legend()

plt.subplot(1, 2, 2)
for i, mf in enumerate(output_mfs):
    plt.plot(x, mf, label=f'Output Set {i+1}')

plt.title('Chiqish mansublik funktsiyalari ')
plt.xlabel('Qiymat')
plt.ylabel('Mansublik')
plt.legend()

plt.show()
```
